## Task 3: Data Transformation (Any Tool)


- Rename the column `availability 365` to `days_booked`
- Convert all column names to lowercase and replace the spaces in the column names with an underscore "_".
- Remove the dollar sign and comma from the columns `price` and `service_fee`. If necessary, convert these two columns to the appropriate data type.

If using Python for this exercise, please include the code in the cells below. If using any other tool, please include screenshoots of your work.

```python
## Rename the column.

nondup.rename({'availability 365':'days_booked'},axis=1).head()
```

```python
## Convert all column names to lowercase and replace the spaces with an underscore "_"

df_convert = nondup.rename({'NAME':'name','host name':'host_name',
       'neighbourhood group':'neighbourhood_group', 'room type':'room_type',
       'Construction year':'Construction_year', 'service fee':'service_fee', 'minimum nights':'minimum_nights',
       'number of reviews':'number_of_reviews', 'last review':'last_review', 'reviews per month':'reviews_per_month',
       'review rate number':'review_rate_number', 'calculated host listings count':'calculated_host_listings_count','availability 365':'days_booked'},axis=1)
```

```python
## Remove the dollar sign and comma from the columns. If necessary, convert these two columns to the appropriate data type.

df_convert.price = df_convert.price.str.replace("$", "").str.replace(",", "").astype(float)
df_convert.service_fee = df_convert.service_fee.str.replace("$", "").str.replace(",", "").astype(float)
```

## Task 5a: Data Visualization (Any Tool)

- Create a horizontal bar chart to display the top 10 most expensive neighborhoods in the dataset.
    - Create another chart with the 10 cheapest neighborhoods in the dataset.
- Create a box and whisker chart that showcases the price distribution of all listings split by room type.

If using Python for this exercise, please include the code in the cells below. If using any other tool, please include screenshoots of your work.

```python
# calculate the median price for each neighbourhood and sort for the 10 most expensive

expneighbourhood = df_convert.groupby("neighbourhood").price.median().round(1).sort_values().tail(10).index
expneighbourhood
```

```python
# Create a horizontal bar chart to display the top 10 most expensive neighborhoods in the dataset.

sns.barplot(y=df_convert.price, x=df_convert["neighbourhood"], order=expneighbourhood)
plt.xticks(rotation=45)
plt.title('Top 10 most expensive neighbourhoods')
```

```python
# # calculate the median price for each neighbourhood and sort for the 10 cheapest

cheapneighbourhood = df_convert.groupby("neighbourhood").price.median().round(1).sort_values().head(10).index
cheapneighbourhood
```

```python
# Create a horizontal bar chart to display the 10 cheapest neighborhoods in the dataset.

sns.barplot(y=df_convert.price, x=df_convert["neighbourhood"], order=cheapneighbourhood)
plt.xticks(rotation=45)
plt.title('10 Cheapest neighbourhoods')
```

```python
# Create a box and whisker chart that showcases the price distribution of all listings split by room type

plt.figure(figsize=(6, 6))
sns.boxplot(x='room_type', y='price', data=df_convert, palette='Set3')
plt.xlabel('room_type')
plt.ylabel('price')
plt.title('Price Distribution by Room Type')
plt.tight_layout()

```

## Task 5b: Data Visualization (Any Tool)

- Create a scatter plot to illustrate the relationshi between the cleaning fee and the room price and write down the kind of correlation, if any, that you see.
- Create a line chart to showcase the total amount of listings available per year.

If using Python for this exercise, please include the code in the cells below. If using any other tool, please include screenshoots of your work.

```python
# Create a scatter plot to illustrate the relationshi between the cleaning fee and the room price and write down the kind of correlation, if any, that you see

# Scatter plot showing relationship between service fee and room price.
plt.figure(figsize=(6, 6))
sns.scatterplot(x=df_convert.price, y=df_convert["service_fee"])
plt.title('Relationship between Service Fee and Room Price')
plt.grid(True)
plt.tight_layout()
```

```python
# Calculate the correlation coefficient

correlation_coefficient = df_convert['service_fee'].corr(df_convert['price'])
print("Correlation Coefficient:", correlation_coefficient)
```

### From the above correlation coefficient, we have a perfect positive correlation which indicates that increase in the Room price produces a corresponding increase in the service fee.

```python
# Create a line chart to showcase the total amount of listings available per year.

# Calculate the total number of listings per year

listings_per_year = df_convert.groupby('Construction_year')['calculated_host_listings_count'].sum().reset_index()

# Create the line chart using seaborn
plt.figure(figsize=(8, 6))
plt.plot(listings_per_year['Construction_year'], listings_per_year['calculated_host_listings_count'], marker='o', color='blue', linestyle='dashed', linewidth=2)
plt.xlabel('Construction Year')
plt.ylabel('Total Listings')
plt.title('Total Number of Listings Available per Construction Year')
plt.grid(True)
plt.tight_layout()
```

```python
#Create a data visualization of your choosing using one of the review columns in isolation or in combination with another column
# Scatter plot showing relationship between reviews per month and minum nights.
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_convert["reviews_per_month"], y=df_convert["minimum_nights"], hue=df_convert.neighbourhood_group)
plt.title('Relationship between Reviews per month and The minimum number of nights stayed')
plt.grid(True)
plt.tight_layout()
```

```python
#Create a visualization to compare at least two different variables between verified hosts identity and unconfirmed hosts identity
#check column names to identify the variables to use 
df_convert.columns
```

```python
# Calculate the average days booked and average minimum nights booked for each host name
comp_data = df_convert.groupby('host_identity_verified').agg({'days_booked': 'mean', 'minimum_nights': 'mean'}).round(1).reset_index()
comp_data
```

```python
# Create the grouped bar chart 

plt.figure(figsize=(8, 6))
sns.barplot(x='host_identity_verified', y='days_booked', data=comp_data, color='orange', label='days_booked')
sns.barplot(x='host_identity_verified', y='minimum_nights', data=comp_data, color='green', label='minimum_nights')

plt.xlabel('Host Name')
plt.ylabel('Average Value')
plt.title('Comparison of days booked and Minimum Nights between Verified and Unconfirmed identity of Hosts')
plt.legend()
plt.grid(True)
plt.tight_layout()
```
